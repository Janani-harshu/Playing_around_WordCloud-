```python
import numpy as np
from PIL import Image

CR7 = open('CR7.txt', 'r').read()

CR7_mask = np.array(Image.open('/content/CR7.jpg'))

# instantiate a word cloud object
CR7_wc = WordCloud(background_color='white', max_words=5000, mask=CR7_mask, stopwords=stopwords,contour_width=5, contour_color='firebrick',color_func=lambda *args, **kwargs: "black", width=2000, height=2000)

CR7_wc.generate(CR7)
fig = plt.figure()
fig.set_figwidth(10) # set width
fig.set_figheight(12) # set height

# display the word cloud
plt.imshow(CR7_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
